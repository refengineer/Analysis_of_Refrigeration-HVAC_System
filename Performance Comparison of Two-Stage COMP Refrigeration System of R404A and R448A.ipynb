# R404A와 R448A의 2단 압축 1단 팽창식 냉동시스템 성능 비교

## Two-Stage Compression Refrigeration System

- ⠀냉동사이클은 압축기에서 일량을 소비하여 증발기로부터 저온 열량을 받아 응축기로 고온 열량을 배출하는 사이클을 말한다. 냉동사이클은 일량을 소비하여 그 결과로 발생되는 열을 이용하는 것으로서, 이때 공급한 일에 대한 이용 열량의 비를 성적계수(COP)라고 한다. 성적계수가 클수록 그 효율은 높다는 것을 의미한다.  
⠀⠀
⠀⠀
- ⠀2단압축 냉동사이클은 증발기에서 주위로부터 열을 빼앗아 증발한 냉매증기를 저단 압축기에서 1차적으로 압축한 다음 고단 압축기에서 다시 한번 압축하여 응축기로 보내는 사이클이다. 여기서 두 압축기 사이에는 중간냉각기를 두고 있다. 중간 냉각기는 저단압축기 토출가스의 냉각과 함께 증발기로 공급하는 액을 과냉각시키는 역할을 함으로써 냉동효과를 증대시키고, 고단압축기에서의 토출가스 온도를 낮추는 데 기여하며, 고단압축기로 흡입되는 가스 중에 있는 액을 분리시키는 역할도 한다. 이러한 중간냉각기에서의 냉각은 냉각수로 할 수도 있지만, 이 사이클에서는 중간냉각기용 팽창밸브에서 팽창되는 냉매액으로 냉각하는 방법을 사용한다. 이렇게 함으로써 응축기에서 응축된 냉매액은 중간냉각기에 설치된 배관을 통과하면서 더욱 과냉각된 후 주 팽창밸브를 통과하여 증발기로 들어가 냉동작용을 하는 사이클이 형성된다.

## Purpose of Analysis
 - ⠀본 내용은 산업용 2단 압축 1단 팽창 냉동시스템에서 주로 쓰이는 R404A냉매와 현시점에서의 친환경적 대체냉매인 R448A냉매를 운전변수에 대한 냉동장치의 성능을 이론적으로 비교 분석해보았다.냉동냉장 솔루션에서 Low GWP 대안 냉매 선정 시 동일 압축기 기준으로 용량 손실이 없는지, 현재 효율과 차이가 나는지, 낮거나 중간 수준의 GWP인지, 가연성 또는 비가연성인지 등의 요소는 반드시 고려해야 한다. HFO 혼합냉매인 R448A는 현재 사용 중인 오존층을 파괴하거나 높은 지구온난화지수를 가진 기존 냉동냉장용 냉매에 대한 규제를 만족하면서 친환경적이고 효율적인 냉매로 인정받고 있다. 특히, 기존 냉매들과 용량은 유사하면서 R404A 대비 약 5~15%의 에너지 효율성이 좋고 비가연성이라는 장점이 있다. 상업용으로 주로 사용되고 있는 R404A와 R507A는 GWP지수가 4000에 육박하고 있어 향후 규제될 가능성이 높다. 하지만 R448A의 경우 GWP가 1300 정도로 HFC 냉매 대비 3분의 1수준으로 향후 규제가 진행되더라도 장기적으로 사용할 있다. 분석을 통해 R404A 및 R448A냉매 적용 시 운전변수에 대한 성능을 비교하여 냉매의 특성을 파악한다.

## Conclusion

- ⠀냉매 R404A와 R448A를 2단 압축 1단 팽창 냉동시스템에 적용시켜 성능을 분석한 결과 모든 조건에서 R448 냉동시스템이 R404A 냉동시스템보다 COP가 높았다.  
⠀⠀
⠀⠀
- ⠀R448A 냉동시스템이 R404A 냉매보다 COP가 높은 이유는 R448A냉매의 습증기 구역이 R404A냉매보다 월등히 크기 때문이다. 이로 인해 R448A냉매는 R404A와 같은 조건에서 비교했을 때 냉동효과 크게 나타나기 때문에 COP가 높다.  
⠀⠀
⠀⠀
- ⠀R448A 냉매는 R404A의 대체 냉매로서 매우 적절하다고 평가할 수 있다. 하지만 HFOs 냉매의 특성상 가격이 매우 높기 때문에 경제성 평가 지표로서는 적절하지 않다.
