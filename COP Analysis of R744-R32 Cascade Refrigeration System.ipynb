# R744-R32용 캐스케이드 냉동시스템의 COP 분석

## CasCade Refrigeration System
- ⠀현재 식품 산업에서 냉동식품이 차지하는 비중은 매우 높으며, 전 세계적으로 냉동식품 시장 규모는 매년 증가하고 있다. 해외 조사 자료에 따르면 2000년대 이후 전 세계 냉동식품 시장은 매년 평균 3.7%의 신장세를 보이고 있으며, 우리나라의 냉동식품 시장도 평균 4.2%로 크게 증가하는 추세이다. 냉동식품을 보관하기 위해서는 대형냉동창고나 대형할인마트에 저온용 냉동시스템이 필요하다. 이러한 냉동시스템은에는 여러 대의 압축기를 이용하여 증발압력에서 응축압력까지 냉매 압력을 단계적으로 올리는 다단 증기 압축식 냉동사이클이 주로 사용되어 왔다. 그러나 다단 증기 압축식 냉동장치는 주로 하나의 단일냉매를 사용하기 때문에 보다 낮은 저온을 얻는데 한계가 있다. 또한, 높은 압축비로 인해 저열원에서 고열원으로 열을 보내는데 압축기에서 소비되는 에너지가 많다. 따라서 이러한 이유로 저온냉동냉장설비용 냉동시스템으로는 독립된 2개의 저온과 고온 사이클이 서로 열교환하여 -50℃의 저온을 얻는 캐스케이드 냉동사이클을 사용하는 것이 유리하다.


## Purpose of Analysis
- ⠀본 내용은 내부 열교환기를 가지는 R744와 R32용 캐스케이드 냉동시스템의 운전변수에 대한 최적의 설계를 위해서 냉동장치의 성능을 이론적으로 분석하였다. 냉매 R744와 R32의 특징으로는 현 시점에서 매우 친환경적인 냉매라고 볼 수 있다. 1971년 14GT이었던 전 세계 온실가스 배출량은 2010년에 이르러 30GT으로 추정되었다. 온실가스가 40년 동안 두 배 이상 급증했다. 과거에 없었던 엄청난 오염원이 쏟아진 꼴이다. 국제에너지기구IEA는 우려한다. 이대로 가면 2050년 무렵 온실가스 배출량이 55GT에 달해, 지구 평균기온이 6℃ 가량 상승할 것으로 예상된다. 냉매관점으로 온실가스를 비교할 때 사용되는 지표는 GWP이다. GWP는 지구온난화지수를 의미하며, 이산화탄소를 GWP=1를 기준할 때 현재 가정에서 사용되는 냉매의 평균 GWP는 대략 2000정도이다. 산업용은 GWP가 무려 4000에 가깝다. 우리는 온실가스를 줄이기 위해 구체적인 해결방안이 필요하다.

### Engineering Equation Solver의 주요 코드  


"!High VCR Cycle"

"Input"



x_High_e_o=1

x_High_c_o=0

⠀⠀

"!Compressor"

P[3]=P[4]  

h_3_ID=enthalpy(RHigh$,P=P[3],s=s[2]) 

W_High_c_ID=(h_3_ID-h[2]) 

W_High_c=W_High_c_ID/(eta_High_c)  

W_High=(W_High_c*m_dot_High)/eta_High_m

h[3]=h[2]+W_High_c    

T[3]=temperature(RHigh$,h=h[3],P=P[3])

s[3]=entropy(RHigh$, P=P[3], h=h[3])

⠀⠀

"!Condenser"

Q_High_Con=h[3]-h[4]   

Q_High_c=Q_High_Con*m_dot_High

⠀⠀

"!Valve"

h[6]=h[5]    

s[6]=entropy(RHigh$,h=h[6],P=P[6])

⠀⠀

"!Evaporator"

P[6]=P[1] 

T[6]=temperature(RHigh$, h=h[6], P=P[6])  

Q_High_Evap=h[1]-h[6] 

Q_High_e=m_dot_High*Q_High_Evap 

COP_High=abs(Q_High_e/W_High)

COP_High_H=abs(Q_High_c/W_High)   

⠀⠀

"!Internal Heat Exchanger"

P[2]=P[1]

eta_High_IHX=(T[2]-T[1])/(T[4]-T[1])

⠀⠀

h[2]-h[1]=h[4]-h[5]

h[2]=enthalpy(RHigh$, T=T[2], P=P[2])

s[2]=entropy(RHigh$, h=h[2], P=P[2])

⠀⠀

Q_High_IH=h[2]-h[1]

Q_High_IHX=m_dot_High*Q_High_IH

⠀⠀

T[5]=temperature(RHigh$, h=h[5], P=P[5])

P[4]=P[5]   

s[5]=entropy(RHigh$, h=h[5], P=P[5]) 

⠀⠀

"!Low VCR cycle"

"Input"

⠀⠀

x_Low_e_o=1

x_Low_c_o=0

⠀⠀

"!Compressor"

P[13]=P[14]  

h_13_ID=enthalpy(RLow$,P=P[13],s=s[12]) 

W_Low_c_ID=(h_13_ID-h[12]) 

W_Low_c=W_Low_c_ID/(eta_Low_c)  

W_Low=(W_Low_c*m_dot_Low)/eta_Low_m

W_Total=W_High+W_Low 

h[13]=h[12]+W_Low_c    

T[13]=temperature(RLow$,h=h[13],P=P[13])

s[13]=entropy(RLow$, P=P[13], h=h[13]) 

⠀⠀

"!Condenser"

Q_Low_Con=h[13]-h[14]    

Q_Low_c=Q_Low_Con*m_dot_Low

⠀⠀

"!Valve"

h[16]=h[15]    

s[16]=entropy(RLow$,h=h[16],P=P[16])

⠀⠀

"!Evaporator"

P[16]=P[11] 

T[16]=temperature(RLow$, h=h[16], P=P[16])  

Q_Low_Evap=h[11]-h[16]   

Q_Low_e=m_dot_Low*Q_Low_Evap   

COP_Low=abs(Q_Low_e/W_Low)  

COP_Low_H=abs(Q_Low_c/W_Low)   

⠀⠀

"!Internal Heat Exchanger"

P[12]=P[11]

eta_Low_IHX=(T[12]-T[11])/(T[14]-T[11])

⠀⠀

h[12]-h[11]=h[14]-h[15]

h[12]=enthalpy(RLow$, T=T[12], P=P[12])

s[12]=entropy(RLow$, h=h[12], P=P[12])

⠀

Q_Low_IH=h[12]-h[11]

Q_Low_IHX=m_dot_Low*Q_Low_IH
 
⠀⠀

T[15]=temperature(RLow$, h=h[15], P=P[15])

P[14]=P[15]   

s[15]=entropy(RLow$, h=h[15], P=P[15]) 

⠀⠀

"!Cascade heat exchanger and VCR total COP"

Q_Low_c=Q_High_e

COP_T=abs(Q_Low_e/(W_High+W_Low))

m_dot_ratio=m_dot_High/m_dot_Low

deltaT_cas=T_Low_c-T_High_e

- ⠀본 데이터 분석을 기반으로 R744-R32용 최적의 캐스케이드 냉동시스템을 설계하기 위해서는 동결실의 증발온도는 높고 실외기의 응축온도가 낮을 때 COP가 증가한다.그 이유는 냉동사이클의 증발압력에 대한 응축압력의 비가 낮아져 압축기의 부하가 작아지기 때문이다.  
⠀⠀
⠀⠀
- ⠀고단 측 및 저단 측의 과냉각도를 증가시키면 COP가 증가한다. 그 이유는 냉매가 과냉각됨에 따라 증발기 내부의 냉매량이 증가하여 주변의 열을 더 쉽게 흡수할 있기 때문이다. 따라서 압축기의 부하는 작아지고 COP가 증가한다. 하지만 본 시스템에서 과열도를 증가시키면 COP가 감소한다. 그 이유는 냉동사이클을 이상적인 과정이라고 가정할 때, 냉매는 압축기에서 단열압축과정을 거치고 등엔트로피의 기울기에 따라 엔탈피가 증가하게 된다. 이때 과열도가 높아짐에 따라 등엔트로피의 기울기는 0의 방향으로 가까워지고 압축기의 소요동력이 상승하여 COP가 감소한다. 냉매는 열역학적 특성이 각각 다르기 때문에 과열도가 높아질 때 COP가 커지는 경우도 있다.  
⠀⠀
⠀⠀
- ⠀본 시스템에서는 내부열교환기 효율을 줄이는 것이 최적의 설계 방법이다. 냉매의 열역학적 물성치에 따라 열교환기 효율이 증가할 때 압축기의 소요동력이 상승하여 COP가 낮아진다. 즉, 효율이 증가될 때 냉동기의 소비전력이 상승됨을 알 수 있다.
